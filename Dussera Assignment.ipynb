## object:- 
An object is a fundamental concept in Python, and it represents a specific instance of a data structure or a class. Objects can have attributes and methods, and they are the building blocks of Python programs.The attributes and methods can be explained:-

### Attributes:-
Attributes are variables associated with the object. These attributes store information about the object. For example, if you have an object representing a person, you could have attributes like name, age, and address.

### Methods:-
Objects can have methods, which are functions associated with the object. Methods can perform actions or operations on the object's data. For example, an object representing a car might have methods like start_engine() and stop_engine().

### Identity and Type: 
Each object in Python has a unique identity, which you can get using the id() function. Objects also have a type, which is determined by their class.

```python
# Creating an object of the built-in int class
x = 5
print(type(x))  # Output: <class 'int'>

# Creating an object of a user-defined class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Creating an instance of the Person class (creating an object)
person1 = Person("John", 30)
```

## rfind()
Python String rfind() method returns the rightmost index of the substring if found in the given string. If not found then it returns -1. Parameters:

sub: It’s the substring that needs to be searched in the given string.

start: Starting position where the sub needs to be checked within the string.

end: Ending position where suffix needs to be checked within the string.

Note: If start and end indexes are not provided then, by default it takes 0 and length-1 as starting and ending indexes where ending indexes are not included in our search.

Return: Returns the right-most index of the substring if it is found in the given string; if not found, then it returns -1.

Syntax: str.rfind(sub, start, end)

```python
string = "oldisgold"
print(string.rfind("is"))

word = 'my name is python'
 
result = word.rfind('name') # Returns highest index of the substring
print ("Substring 'name' found at index :", result )
 
result = word.rfind('python')
print ("Substring 'python' found at index :", result )
 
word = 'CatBatSatMatGate'
 
result = word.rfind('ate') # Returns highest index of the substring
print("Substring 'ate' found at index :", result)
```

Parameters: length: length is the length of the returned string from zfill() with ‘0’ digits filled to the leftside.

Return: Returns a copy of the string with ‘0’ characters padded to the left side of the given string.

Syntax: str.zfill(length)

```python
text = "yummy jimmy kingkong"
  
print(text.zfill(25)) 
  
print(text.zfill(23)) # Given length is less than the length od original string 
  
print(text.zfill(10))
```
