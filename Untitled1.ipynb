## <font color='Red'> Functions:- <font>

In Python, a function is a reusable block of code that performs a specific task or set of tasks. Functions allow you to encapsulate a series of statements into a single unit, which can be called multiple times with different inputs or arguments. Functions make your code more organized, modular, and easier to maintain.

#### syntax :


- Here:
- def: A keyword used to define a function.
- function_name: Name of the function.
- parameters: Optional inputs the function can accept, enclosed in parentheses and separated by commas.

```python
def foosum(a, b):
    print(a + b)
```

```python
foosum(1, 3)
```

### <font color='Green'> Arguments: <font>

- Information can be passed into functions as arguments.
- Arguments are specified after the function name, inside the parentheses.
- You can add as many arguments as you want, just separate them with a comma.
- Python supports 3 types of arguments:

1) Positional Arguments.
2) Keyword Arguments.
3) Default Arguments.


#### Positional Arguments:-

Positional arguments in Python are the most common type of function arguments. They are matched to function parameters in the order in which they are passed. The position (order) of the arguments matters because the function uses this order to assign values to its parameters.

```python
def foodivision(n,d):
    print(n/d)
```

```python
foodivision(10,2)
```

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

message = greet("Alice")
print(message)
```

### <font color='green'> Keyword Argument<font>

Keyword arguments in Python allow you to pass arguments to a function by explicitly specifying the parameter names, rather than relying on the order of positional parameters. This can make your code more readable and help avoid mistakes when calling functions with many parameters.

```python
def foodivision(n,d):
    print(n/d)
```

```python
foodivision(n=2,d=10)
```

```python
def foo(a,b,c):
    x=a+b
    y=x/c
    print(y)
```

```python
foo(1,2,c=5) 
```

```python
foo(1,2,a=5)
```

```python
def greet(name, greeting):
    return f"{greeting}, {name}!"

message = greet(name="Cricket", greeting="Hello")
print(message)  
```

### <font color='Green'> Defualt Argument:<font>

Default arguments in Python allow you to provide a default value for a function parameter. When a default value is specified, you can omit that parameter when calling the function, and the default value will be used instead. This is useful when you want to make a parameter optional and provide a sensible default value in case it's not provided.
