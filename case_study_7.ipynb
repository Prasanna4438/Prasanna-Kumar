## 18.What is the total transaction “penetration” for each product? (hint: penetration = number of transactions where at least one quantity of a product was purchased divided by total number of transactions)

```python
transactions_with_product = df[df['qty'] > 0].groupby('prod_id')['txn_id'].nunique().reset_index()
total_transactions = df['txn_id'].nunique()

penetration_data = pd.merge(transactions_with_product, df[['prod_id', 'product_name']], on='prod_id')
penetration_data['penetration'] = (penetration_data['txn_id'] / total_transactions) * 100
```

## 19.What is the most common combination of at least 1 quantity of any 3 products in a single transaction?

```python
from itertools import combinations
from collections import Counter

valid_transactions = df[df['qty'] > 0]

# Group transactions by 'txn_id' and collect the set of products in each transaction
transaction_products = valid_transactions.groupby('txn_id')['product_name'].apply(set).reset_index()

# Get all combinations of at least one quantity of any 3 products in a single transaction
all_combinations = []
for products_set in transaction_products['product_name']:
    if len(products_set) >= 3:
        combinations_3 = list(combinations(products_set, 3))
        all_combinations.extend(combinations_3)

# Find the most common combination
most_common_combination = Counter(all_combinations).most_common(1)


```

```python

sns.set(style="darkgrid")

# Create subplots with 3 rows and 3 columns
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Histogram for 'price_x'
axes[0, 0].hist(df['price_x'], bins=30, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Histogram of Product Prices')
axes[0, 0].set_xlabel('Product Price')
axes[0, 0].set_ylabel('Frequency')

# Box plot for 'price_x'
sns.boxplot(x=df['price_x'], ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Box Plot of Product Prices')
axes[0, 1].set_xlabel('Product Price')

# Kernel Density Plot for 'price_x'
sns.kdeplot(df['price_x'], ax=axes[0, 2], color='green', fill=True)
axes[0, 2].set_title('Kernel Density Plot of Product Prices')
axes[0, 2].set_xlabel('Product Price')

# Bar chart for 'category_name'
df['category_name'].value_counts().plot(kind='bar', ax=axes[1, 0], color='purple')
axes[1, 0].set_title('Bar Chart of Product Counts by Category')
axes[1, 0].set_xlabel('Category Name')
axes[1, 0].set_ylabel('Count')

sns.lineplot(x=df['discount'], y=np.arange(len(df)), ax=axes[1, 2], color='orange')
axes[1, 2].set_title('Line Chart of Discount')
axes[1, 2].set_xlabel('Discount')
axes[1, 2].set_ylabel('Index')

# Pie chart for 'segment_name'
df['segment_name'].value_counts().plot(kind='pie', ax=axes[1, 1], autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightblue'])
axes[1, 1].set_title('Pie Chart of Product Distribution by Segment')

# Violin plot for 'price_x' with 'category_name'
sns.violinplot(x='category_name', y='price_x', data=df, ax=axes[2, 0], palette='viridis')
axes[2, 0].set_title('Violin Plot of Product Prices by Category')
axes[2, 0].set_xlabel('Category Name')
axes[2, 0].set_ylabel('Product Price')

# Cumulative Distribution Function (CDF) Plot for 'price_x'
sns.ecdfplot(data=df['price_x'], ax=axes[2, 1])
axes[2, 1].set_title('Cumulative Distribution Function (CDF) of Product Prices')
axes[2, 1].set_xlabel('Product Price')
axes[2, 1].set_ylabel('Cumulative Probability')

# Remove empty subplot
fig.delaxes(axes[2, 2])

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
