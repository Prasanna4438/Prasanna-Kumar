### Find()

In Python, the 'Find()' method is used to locate the first occurrence of a substring within a string. It returnsthe index of the first occurrence of the substring in the given string. If substring not found then, it returns -1.


## index()

In python , the 'index()' method is used to find the first occurence of a substring within a string. It is similar to the 'find()' method, but it raises a 'valueerror' if the substring is not found, instead of  returnig -1 as in the case of find()


### Difference Between find() and index()


Error Handling:

index(): Requires you to handle exceptions when the substring is not found, which can be useful when you need to distinguish between a found substring and not found substring.

find(): Does not raise exceptions for not found substrings, making it more suitable for cases where you simply want to check for the existence of the substring.

Return Values:

index(): Returns the index (position) of the first occurrence of the substring within the string or raises a ValueError if not found.

find(): Returns the index (position) of the first occurrence of the substring within the string, or -1 if not found.








### rindex()

The rindex() method is similar to the index() method in Python, but it searches for the last (rightmost) occurrence of a substring within a string. It returns the index (position) of the last occurrence of the substring in the given string. If the substring is not found, it raises a ValueError.

### replace()

In Python, the 'replace()', method is used to replace all occurrences of a specified substring with another substring within a string. The method returns a new string with the replacements made, leaving the original string unchanged.

 ### split()

The split() method splits a string at the specified separator and returns a list of substrings
syntax - str.split(separator, maxsplit)

- separator (optional)- Delimiter at which splits occur. If not provided, the string is splitted at whitespaces.
- maxsplit (optional) - Maximum number of splits. If not provided, there is no limit on the number of splits.

### rsplit()


In Python, the rsplit() method is similar to the split() method, but it splits a string from the right (end) to the left, starting with the last occurrence of the specified separator. The rsplit() method is often used when you want to split a string into substrings starting from the end.

### partition()

The partition() method splits the string into three parts:

The part before the first occurrence of the separator.
The separator itself.
The part after the first occurrence of the separator.
It returns a tuple containing these three parts.

### rpartition()

The rpartition() method is similar to partition(), but it splits the string into three parts based on the last occurrence of the separator.

It splits the string into:

The part before the last occurrence of the separator.
The separator itself.
The part after the last occurrence of the separator.
It returns a tuple containing these three parts.

Both partition() and rpartition() are convenient when you need to extract specific parts of a string based on the positions of separators. The difference between them lies in whether they split based on the first or last occurrence of the separator within the string.

### splitline()


In Python, the splitlines() method is used to split a string into a list of substrings based on line breaks. It is particularly useful when you have a multiline string and want to extract individual lines as separate elements in a list.The splitlines() method is handy for processing text data that is structured into lines, such as reading text files or dealing with multiline strings.
