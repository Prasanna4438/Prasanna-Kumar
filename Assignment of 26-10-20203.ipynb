### List():-
A list is a data structure in Python that is a mutable, or ordered sequence of elements. Each element or value that is inside of a list is called an item. Just as strings are defined as characters between quotes, lists are defined by having values between square brackets [ ]. Lists are Hetrogenous and dynamic in nature.

```python
x = [1, 2.2, 3+3j, 'four', True, print, range]
print(x, type(x), id(x), sep = '\n')
```

## Range()

#### Range:
The range() function is particularly useful when you need to generate a sequence of numbers for iteration, creating lists, or other purposes. It is memory-efficient because it generates numbers on-the-fly rather than creating an entire list of numbers in memory.

### syntax for range:


- start: The starting number of the sequence. If not provided, it defaults to 0.
- stop: The number that defines the end of the sequence. The range() function generates numbers up to, but not including, this value.
- step: The step size, which determines the difference between each number in the sequence. If not provided, it defaults to 1.
- Here start and step are the optional in demiontioning the range.

```python
r = range(1, 11)
print(r, type(r), hex(id(r)), sep = '\n')
```

```python
list(r)
```

```python
R = range(10, 1, -1)
print(r, type(r), hex(id(r)), sep = '\n')
```

```python
list(R)
```

```python
list(range(0,11,1)),list(range(11))
```

### append()
Only one element at a time can be added to the list by using the append() method.
Tuples can also be added to the list with the use of the append method because tuples are immutable.
Unlike Sets, Lists can also be added to the existing list with the use of the append() method.
The List append() method is used for appending and adding elements to the end of the Python List.

#### syntax :


##### Here:
- list: The list you want to add an element.
- element: The element you wanted to add to the list.

```python
z = []
print(z, type(z), len(z),hex(id(z)), sep = '\n')
```

```python
z.append(45)
print(z, type(z), len(z), hex(id(z)), sep = '\n')
```

```python
z.append([1, 2, 3])
print(z, type(z), len(z), hex(id(z)), sep = '\n')
```

## extend()

This method is used to add multiple elements at the same time at the end of the list.
Note: append() and extend() methods can only add elements at the end
extends() method iterates over an iterable like string, list, tuple, etc.,
adds each element of the iterable to the end of the Python List, modifying the original list.

#### syntax()


- Here:
- list: This is the list to which you want to add elements from the iterable.
- iterable: This is the iterable whose elements you want to add to the end of the list.

```python
l = [1, 2, 3]
l.extend([4, 5, 6])
print(l)
```

## pop()

pop() is an inbuilt function that removes and returns the last value from the List or the given index value.
pop() acts on index numbers.

#### syntax:


- Here:
- list: This is the list from which you want to remove an element.
- index: This is the index of the element you want to remove. If not specified, it defaults to the last element in the list.
- element: This is the element that was removed and returned.
- Index is an optional parameter in pop.

```python
x = [1, 2, 3, 4]
y = x.pop(3)
print(y)
```
