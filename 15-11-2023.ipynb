#### Module:-
- In Python, Modules are simply files with the “. py” extension containing Python code that can be imported inside another Python Program.

- In simple terms, we can consider a module to be the same as a code library or a file that contains a set of functions that you want to include in your application.

- Always ensure that the module file (e.g., calculator.py) and the file where you import and use the module (e.g., main.py) are in the same directory or in a location where Python can find the module (like in the Python path).
- You can also organize your modules into packages (folders with an __init__.py file). This helps manage multiple related modules together.

### alias:-
- Creating an alias for a module in Python allows you to refer to the module with a different name within your code.
- This can be helpful if the original module name is lengthy or if you want to avoid naming conflicts.
- This technique can be particularly useful when dealing with modules with long names.
- when you need to differentiate between similar modules or prevent namespace clashes.
