```python
def luFence(x):

    iqr = x.quantile(0.75) - x.quantile(0.25)
    lowerFence, upperFence = max(min(x), x.quantile(0.25)-1.5*iqr), min(max(x), x.quantile(0.75)+1.5*iqr)
    return lowerFence, upperFence
```

```python
luFence(df['fare'])
```

```python
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['fare'] < lower_bound) | (df['fare'] > upper_bound)]

print("Outliers:")
print(outliers)

total_outliers = len(outliers)
print(f"Total Outliers: {total_outliers}")
```

```python
outliers_count_by_pclass = outliers.groupby('pclass').size().reset_index(name='outliers_count')
print(outliers_count_by_pclass)
```

```python

df['pclass_mirror'] = np.where(df['pclass'] == 1, 3, np.where(df['pclass'] == 3, 1, df['pclass']))

# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Create the mirrored kernel density plots
sns.kdeplot(df[df['pclass'] == 1]['age'], label='1st Class', shade=True, color='blue')
sns.kdeplot(df[df['pclass'] == 2]['age'], label='2nd Class', shade=True, color='orange')
sns.kdeplot(df[df['pclass'] == 3]['age'], label='3rd Class', shade=True, color='green')

sns.kdeplot(df[df['pclass'] == 1]['age'], label='', shade=True, color='blue')
sns.kdeplot(df[df['pclass'] == 2]['age'], label='', shade=True, color='orange')
sns.kdeplot(df[df['pclass'] == 3]['age'], label='', shade=True, color='green')

# Remove the mirrored passenger class column
df.drop('pclass_mirror', axis=1, inplace=True)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Butterfly Plot: Pclass vs. Age')
plt.legend(title='Passenger Class')

# Show the plot
plt.show()
```
