### Return:-
- The "return" statement in Python is used to exit a function and return a value.
- The statements after the return statements are not executed.
- If the return statement is without any expression, then the special value None is returned.

```python
def foostrfind(s,char):
    x=s
    y=char
    for i,j in enumerate(x):
        if y==j:
            return i
    else:
            return -1
```

```python
foostrfind('holla','a')
```

### Lambda Function:
Lambda functions, also known as anonymous functions, are a way to create small, unnamed functions in Python. They are defined using the lambda keyword.

#### syntax:-


##### Python lambda properties:
- This function can have any number of arguments but only one expression, which is evaluated and returned.
- One is free to use lambda functions wherever function objects are required.
- You need to keep in your knowledge that lambda functions are syntactically restricted to a single expression.
- It has various uses in particular fields of programming, besides other types of expressions in functions.

```python
(lambda x,y:x**y)(2,4) 
```

```python
n=[1,2,3]
(lambda x: sum(x))([1,2,4])
```

### map():-
The map() function in Python is used to apply a given function to all items in an input list (or any other iterable) and return an iterable (usually a list) of the results. 

#### syntax:-


```python
z=map(lambda x: x**2,[2,3,4])
list(z)
```

### filter():-
The filter() function in Python is used to construct an iterator from elements of an iterable for which a function returns true.

#### syntax:-


Here, function is the function that tests if each element of an iterable is true or false, and iterable is the iterable (e.g., a list) that you want to filter.

```python
# filter  only true values will be executed
list(filter(lambda x:True if x%2==0 else False,[1,2,3,4]))
```

```python
z=list(filter(lambda x:True if x in'aeiouAEIOU' else False,'crciket'))
z
```

```python
list(filter(lambda x:True if  x.isalnum() else False,'cricket@5!'))
```

```python
list(filter(lambda x:True if  x.isalnum() is False else False,'cricket@1!'))
```

```python
list(filter(lambda x:True if  x.islower()  else False,'cricKET@5!'))
```

```python
list(filter(lambda x:True if  x.isupper()  else False,'cricKET@5!'))
```

#### reduce():-

The reduce() function is part of the functools module in Python and is used to perform a cumulative reduction on the items of an iterable, applying a binary function in a cumulative way to the items.

#### syntax:-


- Here.
Here, function is the binary function to apply, and iterable is the iterable (e.g., a list) that you want to reduce. Optionally, you can provide an initializer value.

```python
import functools # we have to import functools to use reduce 
functools.reduce(lambda x,y:x+y,[1,2,3,4]) # it adds first two elements then goes to other 
```
