# <center><font color='green'>Dictionary<font><center>

In Python, a dictionary is a built-in data structure that allows you to store and retrieve data in key-value pairs. They are commonly used for fast data retrieval and mapping between keys and values.Dictionaries are highly versatile and useful for various tasks, such as building configuration settings, caching data, and more.

#### Creating a Dictinary:


```python
x = {} #empty dictionary

print(x, type(x), sep = '\n')
```

```python
x = dict()
print(x, type(x), sep = '\n')
```

```python
x = dict(((4,131),(23,2200)))
print(x, type(x), sep = '\n')
```

```python
x = dict(((11,[21, 22, 23, 111]),(4,45)))
print(x, type(x), sep = '\n')
```

```python
x = {2:[11,2.3,44],
    'hi':(25,35,45),
    ('a','b'):50,
    'name':'harshit'}
print(x['hi'])
print(x['hi'][1])
print(x['name'])
print(x['name'][4])
```

#### copy():
You can create a shallow copy of a dictionary using the copy() method. This method copies the key-value pairs to a new dictionary, so changes in the original dictionary won't affect the copy, and vice versa.

```python
a = {'x':11, 'y':22}

b = a.copy()

print(a)
print(b)
```

#### fromkeys():
If you want to extract a value from a dictionary using a key, you can simply access the value associated with that key using the key within square brackets

```python
x = {'a': 111, 'b': 222, 'c': 333, 'd': 444, 'e': 555, 'f': 666, 'g': 777}

print(x.fromkeys(x))
```

#### get():

In Python, you can use the get() method on a dictionary to retrieve a value associated with a key. This method is useful because it allows you to specify a default value to return if the key doesn't exist in the dictionary.

##### syntax:
dictionary.get(keyname, value)

key - key to be searched in the dictionary
value (optional) - Value to be returned if the key is not found. The default value is None.
Return Value

the value for the specified key if key is in the dictionary.
None if the key is not found and value is not specified.
value if the key is not found and value is specified.

```python
x = {'a':111,
     'b':222,
     'c':333}
print(x)
print(x.get('b'))
```

#### items():
In Python, you can use the items() method to retrieve a list of key-value pairs (tuples) from a dictionary. This method returns an iterable view object that provides access to the dictionary's items (key-value pairs). Here's how you can use it:




```python
x = {'a': 111, 'b': 222, 'c': 333, 'd': 444, 'e': 555, 'f': 666, 'g': 777}

print(x)
x.items()
```

#### key():
In Python, you can access the keys of a dictionary using various methods, including using the keys() method, iterating through the dictionary, or converting the keys to a list. Here are some ways to access the keys of a dictionary:

```python
x = {'a1':10,
     'a2':20,
     'a3':30}
print(x)
x.keys()
a = x.keys()
print(a, type(a), len(a), sep = '\n')
```

#### pop():
In Python, the pop() method is used to remove an item from a dictionary based on its key and return the corresponding value. 

dictionary.pop(key[, default])

```python
x = {'a': 111, 'b': 222, 'c': 333, 'd': 444, 'e': 555, 'f': 666, 'g': 777}

print(x.pop('a'))
```

#### popitem():
The popitem() method removes and returns the (key, value) pair from the dictionary in the Last In, First Out (LIFO) order.

Returns the latest inserted element (key,value) pair from the dictionary.
Removes the returned element pair from the dictionary.

```python
x = {'a': 111, 'b': 222, 'c': 333, 'd': 444, 'e': 555, 'f': 666, 'g': 777}

print(x.popitem())
```

#### setdefault():

In Python, the setdefault() method is used to set a default value for a key in a dictionary if the key does not exist. If the key is already in the dictionary, the method returns its current value without modifying the dictionary.

```python
x = {'a1': 'apple',
     'a2': 'mango',
     'a3': 2000}

y = x.setdefault('a2', 'banana')
z = x.setdefault('a3', 2020)
print(y)
print(z)
```

#### update():
In Python, the update() method is used to merge the contents of one dictionary into another dictionary. It takes another dictionary as its argument and updates the calling dictionary by adding the key-value pairs from the provided dictionary. If a key already exists in the calling dictionary, the value is updated with the value from the provided dictionary.

```python
x = {'a':111,
     'b':222,
     'c':333}
print(x)
x.update([('d' , 444),
          ('e' , 555)])
print(x)
```

#### values():
In Python, the values() method is used to retrieve a view of the values in a dictionary. This view can be used to iterate through the values or convert them into a list.

```python
x = {'a1':10,
     'a2':20,
     'a3':30}
print(x)
x.values()
a = x.values()
print(a, type(a), len(a), sep = '\n')
```

### maketrans():
The maketrans() function is a part of Python's str class, and it is used for creating a translation table that can be used with the translate() method to replace specified characters in a string with other characters. It is often used for tasks like text manipulation, character replacement, or character removal. The maketrans() function can be particularly useful for text transformations.You can use maketrans() and translate() for various text manipulation tasks, such as replacing characters, removing characters, or performing more complex translations between characters or substrings.


```python
help(str.maketrans)
```

```python
mt ="welcome to hython"
m = str.maketrans("h", "p")
print(mt.translate(m))
```

## translate():
The translate() method in Python is used to replace characters in a string or remove characters based on a translation table created using the maketrans() method. This method is particularly useful when you need to perform character replacement or removal in a string.

- text.translate(translation_table) applies the translation table to the text and replaces the specified characters based on the mapping.
- You can use the translate() method for various text manipulation tasks, such as character replacement, character removal, and more. It is efficient for performing multiple character replacements within a single string.

```python
help(str.translate)
```
